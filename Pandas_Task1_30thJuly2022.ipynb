```python
import pandas as pd
```

```python
df=pd.read_excel(r'C:\Users\ankush-sharma2\OneDrive - MMC\Documents\iNeuron\Pandas_Dataset\AgentPerformance.xlsx')
```

```python
df.head()
```

```python
#Adding a new converted datetime column
df['Converted_Date'] = pd.to_datetime(df['Date'])
```

```python
#Adding a new converted week column
df['Converted_week'] = df['Converted_Date'].dt.week
```

```python
#Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave
# OR just not taking into account those dates where total chats,Average Response Time,Average Resolution Time,Average Rating is not zero
df2=df[ (df['Total Chats'] > 0) & (df['Average Response Time'] !='00:00:00') & (df['Average Resolution Time'] != '00:00:00') & (df['Average Rating'] != '0.0') & (df['Total Feedback'] != '0.0')]
```

```python
#Calculating average rating of person on basis of weeks excluding the holidays taken
df2.groupby(['Agent Name', 'Converted_week'])['Average Rating'].mean()
```
